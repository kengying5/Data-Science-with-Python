# Normalisation and Scaling

```python
from scipy import stats
from mlxtend.preprocessing import minmax_scaling
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
kickstarter = pd.read_csv('C:\Users\Sim Keng Ying\Desktop\kickstarter-projects\ks-projects-201801.csv')
```

Scaling

```python
usd_goal = kickstarter.usd_goal_real
scaled_data = minmax_scaling(usd_goal, columns = [0])

fig, ax=plt.subplots(1,2)
sb.distplot(kickstarter.usd_goal_real, ax=ax[0])
ax[0].set_title("Original Data")
sb.distplot(scaled_data, ax=ax[1])
ax[1].set_title("Scaled data")
```

Normalisation

```python
index_of_positive_pledges = kickstarter.usd_pledged_real > 0
index_of_positive_pledges
```

```python
positive_pledges = kickstarter.usd_pledged_real.loc[index_of_positive_pledges]
positive_pledges
```

```python
normalized_pledges = stats.boxcox(positive_pledges)[0]

fig, ax=plt.subplots(1,2)
sb.distplot(positive_pledges, ax=ax[0])
ax[0].set_title("Original Data")
sb.distplot(normalized_pledges, ax=ax[1])
ax[1].set_title("Normalized data")
```

# Encoding and Inconsistent Data Entries

```python
import fuzzywuzzy
from fuzzywuzzy import process
import chardet
```

```python
pakistan = pd.read_csv('C:\Users\Sim Keng Ying\Desktop\PakistanSuicideAttacks.csv')
pakistan.head()
```

```python
cities = pakistan['City'].unique()
cities.sort()
cities
```

```python
pakistan['City'] = pakistan['City'].str.lower()
pakistan['City'] = pakistan['City'].str.strip()
cities = pakistan['City'].unique()
cities.sort()
cities
```

```python
matches = fuzzywuzzy.process.extract("d.i khan", cities, limit=10, scorer=fuzzywuzzy.fuzz.token_sort_ratio)
matches
```

```python
def replace_matches_in_column(df, column, string_to_match, min_ratio = 90):
    
    strings = df[column].unique()
    
    matches = fuzzywuzzy.process.extract(string_to_match, strings, 
                                         limit=10, scorer=fuzzywuzzy.fuzz.token_sort_ratio)

    
    close_matches = [matches[0] for matches in matches if matches[1] >= min_ratio]

    
    rows_with_matches = df[column].isin(close_matches)
 
    df.loc[rows_with_matches, column] = string_to_match
    
    print("Done.")
```

```python
replace_matches_in_column(df=pakistan, column='City', string_to_match="d.i khan")
```
