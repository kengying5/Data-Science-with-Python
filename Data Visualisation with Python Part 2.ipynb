```python
address = 'C:\Users\Sim Keng Ying\Desktop\Ex_Files_Python_Data_Science_EssT\mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
mpg = cars['mpg']
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
mpg.plot(kind='hist', color = 'salmon')
ax.set_xticks(range(35))
```

```python
sb.distplot(mpg, color = 'powderblue')
```

```python
cars.plot(kind='scatter', x='hp', y='mpg', c=['powderblue'], s=150)
```

```python
sb.regplot(x='hp', y='mpg', data=cars, scatter=True)
```

```python
sb.pairplot(cars)
```

```python
cars_df = pd.DataFrame((cars.ix[:,(1,3,4,6)].values), columns = ['mpg', 'disp', 'hp', 'wt'])
cars_target = cars.ix[:,9].values
target_names = [0, 1]

cars_df['group'] = pd.Series(cars_target, dtype="category")
sb.pairplot(cars_df, hue='group', palette='hls')
```

```python
cars.boxplot(column='mpg', by='am')
```
